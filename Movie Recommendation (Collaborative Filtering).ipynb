```python
import pandas as pd
import numpy as np
from sklearn.metrics.pairwise import cosine_similarity
```

```python
ratings=pd.read_csv("dataset/ratings.csv")
movies=pd.read_csv("dataset/movies.csv")
```

```python
ratings.head()
```

```python
movies.head()
```

```python
ratings=pd.merge(movies,ratings)
```

```python
ratings
```

```python
ratings.drop(['genres','timestamp'],axis=1)
```

```python
ratings=ratings.pivot_table(index=['userId'],columns=['title'],values='rating')
```

```python
ratings
```

```python
ratings=ratings.dropna(thresh=10,axis=1).fillna(0)
```

```python
ratings
```

```python
item_similarity_df=ratings.corr(method='pearson')
```

```python
item_similarity_df
```

```python
def find_similarity(name,user_rating):
    similar_scores = item_similarity_df[name]*(user_rating-2.5)
    similar_scores=similar_scores.sort_values(ascending=False)
    return similar_scores
```

```python
action_lover = [("(500) Days of Summer (2009)",1),("Zodiac (2007)",5)]

similar_movies = pd.DataFrame()
for movie,rating in action_lover:
    similar_movies=similar_movies.append(find_similarity(movie,rating))
similar_movies.sum().sort_values(ascending=False)
```

```python
item_similarity_df.to_csv("item_similarity_df.csv")
```
